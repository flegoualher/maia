```python
import os
import shutil

def move_laz_files(source_folder, destination_folder):
    # Create the destination folder if it doesn't exist
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Get a list of files in the source folder
    files = os.listdir(source_folder)

    # Move only .laz files to the destination folder
    for file in files:
        if file.lower().endswith(".laz"):
            source_path = os.path.join(source_folder, file)
            destination_path = os.path.join(destination_folder, file)
            shutil.move(source_path, destination_path)
            print(f"Moved: {file}")

if __name__ == "__main__":
    source_folder = "/Users/user/Library/Mobile Documents/com~apple~CloudDocs/[02]-work/[02]-2023-le-wagon/[04]-final-project/MX19_Inomata_complete"
    destination_folder = "/Users/user/Library/CloudStorage/GoogleDrive-fredericlgtx@gmail.com/My Drive/MX19_Inomata/point_cloud_tiles"
    move_laz_files(source_folder, destination_folder)
```
